```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib notebook

import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.DataFrame({"ratio (fake:real)": ["1:1", "1:2", "1:3", "1:4", "1:5", "1:6", "1:7", "1:8", "1:9", "1:22"],
      "precision (fake)": [.70, .72, .68, .77, .73, .68, .77, .69, .77, .79],
      "model accuracy": [.75, .81, .86, .89, .89, .92, .92, .92, .93, .97] })
df
```

```python
# Assign x and y values
x_axis = np.arange(1,len(df)+1,1)
y_values_p = df['precision (fake)']
y_values_a = df['model accuracy']

# Create handles
precision, = plt.plot(x_axis, y_values_p, color="navy", label="Precision Predicting Fake")
accuracy, = plt.plot(x_axis, y_values_a, color="fuchsia", label="Overall Accuracy of Model")

# Plot ticks
tick_locations = [value for value in x_axis]
tick_labels = df['ratio (fake:real)']
plt.xticks(tick_locations,tick_labels)
plt.ylim(.5, 1, .1)

# Add titles, labels, legend
plt.title("Model Precision/Accuracy as Sample of Real Data Increases")
plt.xlabel("Fake:Real Sample Ratio")
plt.ylabel("Precision/Accuracy")

plt.legend(handles=[precision, accuracy], loc="lower right")
plt.grid()
plt.show()
```
