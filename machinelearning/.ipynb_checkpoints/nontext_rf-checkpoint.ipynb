# Load Data

```python
import pandas as pd
pd.set_option('display.max_columns', 30)

import warnings
warnings.filterwarnings("ignore")
```

```python
# Read in csv file
df = pd.read_csv("../assets/data/fake_jobs_clean.csv")
df.head(60)
```

```python
list(df.columns)
```

# Feature Selection

```python
# Feature Selection
selected_features = df.drop(['job_id', 'city', 'state/province', 'country','title','department', 'salary_range', 
                             'company_profile', 'description', 'requirements','benefits', 'fraudulent'], axis=1)
selected_features
```

```python
# Get Dummies for Categorical Columns
dummy_features = pd.get_dummies(selected_features)
dummy_features
```

```python
print(len(list(dummy_features.columns)))
list(dummy_features.columns)
```

```python
# Assign the data to X and y
X = dummy_features
y = df['fraudulent']

print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=22)
```

# Pre-Processing

Scale Data using the MinMaxScaler

```python
# Scale the data using the MinMaxScaler
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from tensorflow.keras.utils import to_categorical

X_scaler = MinMaxScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

# Train the Model

```python
#Random Forest algorithm
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=200, random_state=22)

#Fit the model to the data
rf = rf.fit(X_train_scaled, y_train)

#Quantify the model
rf.score(X_test_scaled, y_test)
```

```python
# Check Feature Importance
sorted(zip(rf.feature_importances_, selected_features), reverse=True)
```

```python
# Classification Model
from sklearn.metrics import classification_report
predictions = rf.predict(X_test)
print("RF Classification Report: \n" + classification_report(y_test, predictions, target_names=['real','fake']))
```

# Hyperparameter Tuning

```python
# Create the GridSearchCV model
from sklearn.model_selection import GridSearchCV
param_grid = { 
    'n_estimators': [200, 500],
    'max_features': ['auto', 'sqrt', 'log2'],
    'max_depth' : [4,5,6,7,8],
    'criterion' :['gini', 'entropy']
}

CV_rf = GridSearchCV(estimator=rf, param_grid=param_grid, cv= 5)
```

```python
# Train the model with GridSearch
CV_rf.fit(X_train_scaled, y_train)
```

```python
#Get best params
print(CV_rf.best_params_)
print(CV_rf.best_score_)
```

```python
#Create new RF model with best params
rf2 = RandomForestClassifier(random_state=22, max_features='auto', n_estimators= 200, max_depth=8, criterion='gini')

#Fit the model to the data
rf2 = rf2.fit(X_train_scaled, y_train)

#Quantify the model
rf2.score(X_test_scaled, y_test)
```

```python
# Classification Model
from sklearn.metrics import classification_report
predictions = rf2.predict(X_test)
print("RF2 Classification Report: \n" + classification_report(y_test, predictions, target_names=['real','fake']))
```

# Save the Model

```python
import joblib
joblib.dump(rf2, 'nontext_rf2.h5')
```
